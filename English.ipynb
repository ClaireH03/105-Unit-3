# Second Step
## Importing County Health Data 
Next, we imported our County Health Data CSV using the `pd.read_csv()` function

```python
df=pd.read_csv("CountyHealthData_2014-2015 (1).csv")
```

# Third Step

Now that the data has been inserted we can now alter the data by using different attributes using the function `<DataFrame name>.<attribute name>`

# Fourth Step
## Extracting Single Columns 
Here we can look at a single column from the data set using the 'Series' function

We can extract a series using bracket notation `df["Children in poverty"]`

```python
df["Children in poverty"]
```

# Fifth Step
## Viewing NC Data Only
Now we will view only the NC data by filtering our data to only show NC using the `df[df["State"] == "NC"]`
function
* the inner statement looks for column with the name NC 
* the outer statement looks for rows that match that column

```python
df[df["State"] == "NC"]
```

# Sixth Step
## Defining The Subset
Now we will define this subset using the `.copy()` function so we can manipulate the data only within this subset

```python
NC_subset = df[df["State"] == "NC"].copy()
```

Now we can recall the data and view only the NC_subset

```python
NC_subset
```

# Seventh Step
## Maipulating the NC data
Now we will view only the columns `["Children in poverty", "Population that is not proficient in English", "Median household income"]`
* We do this by using the `set.loc` function that searches for columns with the set attributes 
* So here we will enter in `NC_subset.loc[:,["Children in poverty", "Population that is not proficient in English", "Median household income"]]`
* the : signifies using viewing every row within the dataset 

```python
NC_subset.loc[:,["Children in poverty", "Population that is not proficient in English", "Median household income"]]
```

# Eight Step
## Defining Our Final Subset
Now we will define our subset with our column filters again using the same protocal in step six
* This Function will be `Final_Subset= NC_subset.loc[:,["Children in poverty", "Population that is not proficient in English", "Median household income"]].copy()`

```python
Final_Subset= NC_subset.loc[:,["Children in poverty", "Population that is not proficient in English", "Median household income"]].copy()
```

```python
Final_Subset
```

# Final Step
## Exporting The Filtered Data in a CSV
Now that we have sucessfully filtered the data we will export it to a .cvs file using `Final_Subset.to_csv("Final_Subset.csv", index=False)`
the `index=False` allows us to get rid of the index number

```python
Final_Subset.to_csv("Final_Subset.csv", index=False)
```
